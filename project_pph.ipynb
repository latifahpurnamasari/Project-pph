## Data cleaning

### Subtask:
Clean the data by handling missing values, converting relevant columns to numeric types, and addressing potential inconsistencies.


**Reasoning**:
Identify and convert 'object' type columns that should be numeric, handle non-numeric characters and errors, then address missing values in all relevant columns. Finally, verify data types and check for remaining missing values.



```
import numpy as np

# Identify columns that should be numeric but are 'object' type
numeric_cols_to_convert = [
    'persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan',
    'total_hpp', 'pendapatan_operasional', 'pendapatan_penjualan',
    'ppn_keluaran', 'ppn_masukan', 'ppn_netto', 'laba_kotor_penjualan',
    'beban_operasional', 'overhead_farmasi', 'penyusutan_total',
    'penyusutan_garis_lurus', 'penyusutan_saldo_menurun', 'penyusutan_fiskal',
    'laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak',
    'pph_badan_normal', 'tarif_pph_tax_holiday', 'pph_badan_tax_holiday',
    'laba_bersih_normal', 'laba_bersih_tax_holiday', 'kerugian_fiskal'
]

# Convert identified columns to numeric, handling non-numeric characters
for col in numeric_cols_to_convert:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert percentage column 'tarif_pph_normal' to numeric
if 'tarif_pph_normal' in df.columns:
    df['tarif_pph_normal'] = df['tarif_pph_normal'].astype(str).str.replace('%', '', regex=False)
    df['tarif_pph_normal'] = pd.to_numeric(df['tarif_pph_normal'], errors='coerce') / 100.0

# Address missing values for numeric columns by filling with the median
for col in numeric_cols_to_convert + ['tarif_pph_normal']:
     if col in df.columns and df[col].isnull().any():
         median_val = df[col].median()
         df[col].fillna(median_val, inplace=True)


# Address missing values in 'tax_holiday_status' by filling with a placeholder
if 'tax_holiday_status' in df.columns and df['tax_holiday_status'].isnull().any():
    df['tax_holiday_status'].fillna('Tidak Diketahui', inplace=True)

# Verify data types after cleaning
display(df.info())

# Check for remaining missing values
display(df.isnull().sum())
```

## Data analysis

### Subtask:
Analyze the cleaned data to derive meaningful insights. This might involve calculating aggregations, identifying trends, or comparing different categories within the data.


**Reasoning**:
Perform the requested data analysis steps: calculate total laba_sebelum_pajak by year, determine average laba_bersih for normal and tax holiday statuses, group by tax_holiday_status to sum laba_sebelum_pajak and laba_kena_pajak, and identify years with highest/lowest laba_sebelum_pajak.



```
# 1. Calculate total 'laba_sebelum_pajak' for each year
total_laba_sebelum_pajak_by_year = df.groupby('tahun')['laba_sebelum_pajak'].sum()
print("Total 'laba_sebelum_pajak' for each year:")
display(total_laba_sebelum_pajak_by_year)

# 2. Determine the average 'laba_bersih_normal' and 'laba_bersih_tax_holiday' across all years
average_laba_bersih_normal = df['laba_bersih_normal'].mean()
average_laba_bersih_tax_holiday = df['laba_bersih_tax_holiday'].mean()
print("\nAverage 'laba_bersih_normal' across all years:")
print(average_laba_bersih_normal)
print("\nAverage 'laba_bersih_tax_holiday' across all years:")
print(average_laba_bersih_tax_holiday)

# 3. Group the data by 'tax_holiday_status' and calculate the sum of 'laba_sebelum_pajak' and 'laba_kena_pajak' for each group
laba_by_tax_holiday_status = df.groupby('tax_holiday_status')[['laba_sebelum_pajak', 'laba_kena_pajak']].sum()
print("\nSum of 'laba_sebelum_pajak' and 'laba_kena_pajak' by 'tax_holiday_status':")
display(laba_by_tax_holiday_status)

# 4. Identify the year with the highest and lowest 'laba_sebelum_pajak'
year_highest_laba_sebelum_pajak = total_laba_sebelum_pajak_by_year.idxmax()
highest_laba_sebelum_pajak_value = total_laba_sebelum_pajak_by_year.max()
year_lowest_laba_sebelum_pajak = total_laba_sebelum_pajak_by_year.idxmin()
lowest_laba_sebelum_pajak_value = total_laba_sebelum_pajak_by_year.min()

print(f"\nYear with the highest 'laba_sebelum_pajak': {year_highest_laba_sebelum_pajak} (Value: {highest_laba_sebelum_pajak_value})")
print(f"Year with the lowest 'laba_sebelum_pajak': {year_lowest_laba_sebelum_pajak} (Value: {lowest_laba_sebelum_pajak_value})")
```

## Data visualization

### Subtask:
Create relevant and visually appealing charts with appropriate colors to represent the findings from the data analysis.


**Reasoning**:
Generate the requested plots to visualize the data analysis findings.



```
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style("whitegrid")

# 1. Bar chart for total 'laba_sebelum_pajak' per year
plt.figure(figsize=(10, 6))
total_laba_sebelum_pajak_by_year = df.groupby('tahun')['laba_sebelum_pajak'].sum()
colors1 = sns.color_palette("viridis", len(total_laba_sebelum_pajak_by_year))
total_laba_sebelum_pajak_by_year.plot(kind='bar', color=colors1)
plt.title('Total Laba Sebelum Pajak per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Laba Sebelum Pajak')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Bar chart for average 'laba_bersih_normal' and 'laba_bersih_tax_holiday'
average_laba_bersih_normal = df['laba_bersih_normal'].mean()
average_laba_bersih_tax_holiday = df['laba_bersih_tax_holiday'].mean()
average_laba_df = pd.DataFrame({
    'Laba Bersih': [average_laba_bersih_normal, average_laba_bersih_tax_holiday]
}, index=['Normal', 'Tax Holiday'])

plt.figure(figsize=(7, 5))
colors2 = sns.color_palette("plasma", 2)
average_laba_df.plot(kind='bar', color=colors2, legend=False)
plt.title('Rata-rata Laba Bersih: Normal vs Tax Holiday')
plt.xlabel('Status Pajak')
plt.ylabel('Rata-rata Laba Bersih')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 3. Grouped bar chart for sum of 'laba_sebelum_pajak' and 'laba_kena_pajak' by 'tax_holiday_status'
laba_by_tax_holiday_status = df.groupby('tax_holiday_status')[['laba_sebelum_pajak', 'laba_kena_pajak']].sum()

plt.figure(figsize=(10, 6))
colors3 = sns.color_palette("tab10", 2)
laba_by_tax_holiday_status.plot(kind='bar', color=colors3)
plt.title('Total Laba Sebelum Pajak dan Laba Kena Pajak berdasarkan Status Tax Holiday')
plt.xlabel('Status Tax Holiday')
plt.ylabel('Total Laba')
plt.xticks(rotation=0)
plt.legend(['Laba Sebelum Pajak', 'Laba Kena Pajak'])
plt.tight_layout()
plt.show()
```

## Data cleaning

### Subtask:
Clean the data by handling missing values, correcting data types for numerical columns, and addressing inconsistencies identified during the exploration phase.


**Reasoning**:
Identify the columns that should be numerical but are currently of 'object' type and clean them by removing non-numeric characters and converting them to numeric. Handle missing values by filling them with the median for numerical columns, as the exploration showed a significant number of missing values and using the median is robust to outliers. Address inconsistencies by removing rows with non-numeric values in columns expected to be numeric, except for the 'tahun' column which contains "Normal" and "22%", where "Normal" seems to be an anomaly and "22%" is a rate that should be handled separately if needed for calculation, but for now, we will focus on cleaning the numerical columns. The 'tarif_pph_normal' column contains '%' and will be converted to float after removing '%'. The 'tahun' column also contains 'Normal', which is not a year and will be removed. The 'tax_holiday_status' column contains 'Ya' and 'Tidak' which are categorical and will be kept as is. The 'penyusutan_garis_lurus' and 'penyusutan_saldo_menurun' columns contain 'Garis Lurus' and 'Saldo Menurun' which are categorical descriptions and will be handled by removing rows where these appear in numerical columns.



```
import numpy as np

# Columns identified as numerical but are 'object' type
numerical_cols_to_clean = [
    'persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan',
    'total_hpp', 'pendapatan_operasional', 'pendapatan_penjualan', 'ppn_keluaran',
    'ppn_masukan', 'ppn_netto', 'laba_kotor_penjualan', 'beban_operasional',
    'overhead_farmasi', 'penyusutan_total', 'penyusutan_garis_lurus',
    'penyusutan_saldo_menurun', 'penyusutan_fiskal', 'laba_kotor_operasional',
    'laba_sebelum_pajak', 'laba_kena_pajak', 'pph_badan_normal',
    'tarif_pph_tax_holiday', 'pph_badan_tax_holiday', 'laba_bersih_normal',
    'laba_bersih_tax_holiday', 'kerugian_fiskal'
]

# Clean and convert numerical columns
for col in numerical_cols_to_clean:
    if col in df.columns:
        # Remove periods, commas, and spaces
        df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False).str.replace(' ', '', regex=False)
        # Convert to numeric, coercing errors to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Clean and convert 'tarif_pph_normal' column
if 'tarif_pph_normal' in df.columns:
    df['tarif_pph_normal'] = df['tarif_pph_normal'].astype(str).str.replace('%', '', regex=False).str.strip()
    df['tarif_pph_normal'] = pd.to_numeric(df['tarif_pph_normal'], errors='coerce')


# Handle inconsistencies in 'tahun' column
if 'tahun' in df.columns:
    # Convert to numeric, coercing errors to NaN for non-year values
    df['tahun'] = pd.to_numeric(df['tahun'], errors='coerce')
    # Drop rows where 'tahun' is NaN after coercion (removing "Normal" and other non-year entries)
    df.dropna(subset=['tahun'], inplace=True)
    # Convert 'tahun' to integer after dropping NaNs
    df['tahun'] = df['tahun'].astype(int)


# Handle missing values by filling with median for numerical columns
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col].fillna(df[col].median(), inplace=True)

# Verify the data types and check for remaining missing values
display(df.info())
display(df.isnull().sum())
```
