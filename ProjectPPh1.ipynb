## Data cleaning

### Subtask:
Clean the data by handling missing values and converting columns with incorrect data types.


**Reasoning**:
Inspect the 'object' columns to identify non-numeric entries and prepare for conversion to numeric types.



```python
for col in df.select_dtypes(include='object').columns:
    if col not in ['tahun', 'tax_holiday_status', 'tarif_pph_normal']: # Exclude known non-numeric columns
        non_numeric_values = df[col][pd.to_numeric(df[col], errors='coerce').isna()].unique()
        if len(non_numeric_values) > 0:
            print(f"Column '{col}' contains non-numeric values: {non_numeric_values}")
```

**Reasoning**:
Convert the identified object columns with non-numeric entries to numeric, handling the specific non-numeric values by replacing them with NaN, then convert to numeric, and finally handle the resulting missing values by dropping rows with more than a certain threshold of NaNs and imputing the rest.



```python
import numpy as np

# Define columns that should be numeric but are object type
numeric_cols_to_convert = [
    'persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan',
    'total_hpp', 'pendapatan_operasional', 'pendapatan_penjualan', 'ppn_keluaran',
    'ppn_masukan', 'ppn_netto', 'laba_kotor_penjualan', 'beban_operasional',
    'laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak',
    'laba_bersih_normal', 'laba_bersih_tax_holiday'
]

# Clean and convert columns
for col in numeric_cols_to_convert:
    # Remove periods and spaces, replace commas with periods for decimal conversion if necessary (although none found in output)
    df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False).str.strip()
    # Replace specific non-numeric strings with NaN
    df[col] = df[col].replace(['Tax Holiday', 'Ya', 'Metode Depresiasi', 'Garis Lurus', 'garis_lurus',
                               'Saldo Menurun', 'saldo_menurun', 'aset_id', 'A001', 'A002', 'A003',
                               'A004', 'A005', 'kategori', 'mesin', 'kendaraan', 'bangunan',
                               'nilai_perolehan', 'umur_ekonomis', 'metode_depresiasi',
                               'penyusutan_garis_lurus', 'penyusutan_saldo_menurun', '-', 'penyusutan_fiskal'], np.nan)
    # Convert to numeric, coercing errors
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert 'tarif_pph_normal' by removing '%' and converting to float
df['tarif_pph_normal'] = df['tarif_pph_normal'].astype(str).str.replace('%', '', regex=False).str.strip()
df['tarif_pph_normal'] = pd.to_numeric(df['tarif_pph_normal'], errors='coerce') / 100

# Handle missing values: Drop rows with more than 50% missing values
threshold = len(df.columns) * 0.5
df.dropna(thresh=threshold, inplace=True)

# Impute remaining missing numerical values with the median
for col in df.select_dtypes(include=np.number).columns:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

# Verify data types and missing values
display("Data types after cleaning and conversion:")
display(df.info())
display("Missing values after cleaning and imputation:")
display(df.isnull().sum())
```

## Data wrangling

### Subtask:
Prepare the data for analysis and visualization. This involves ensuring the 'tahun' column is in a suitable format and potentially selecting relevant columns for specific visualizations.


**Reasoning**:
Convert the 'tahun' column to datetime objects, select relevant columns for time series analysis, and ensure all selected numerical columns are of a numeric data type.



```python
# 1. Convert the 'tahun' column to a datetime format
df['tahun'] = pd.to_datetime(df['tahun'], format='%Y')

# 2. Select the columns relevant for time-series analysis
# Assuming relevant columns are the year and the numerical columns
relevant_cols = ['tahun'] + df.select_dtypes(include=np.number).columns.tolist()
df_ts = df[relevant_cols].copy()

# 3. Ensure the selected numerical columns are of a numeric data type
# This was already done in the previous cleaning step, but we can verify
for col in df_ts.select_dtypes(include=np.number).columns:
    if not pd.api.types.is_numeric_dtype(df_ts[col]):
        df_ts[col] = pd.to_numeric(df_ts[col], errors='coerce')

display(df_ts.head())
display(df_ts.info())
```

## Data analysis

### Subtask:
Perform data analysis to extract insights from the data, focusing on trends over time for key financial metrics.


**Reasoning**:
Calculate the total profit metrics for each year, identify the years with highest and lowest values, calculate the average tax rate, and determine the total PPH Badan Normal.



```python
# Group by year and calculate the sum of profit metrics
profit_by_year = df_ts.groupby(df_ts['tahun'].dt.year)[['laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_bersih_normal']].sum()

display("Total Profit Metrics by Year:")
display(profit_by_year)

# Identify years with highest and lowest values for each metric
for col in ['laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_bersih_normal']:
    max_year = profit_by_year[col].idxmax()
    min_year = profit_by_year[col].idxmin()
    display(f"'{col}':")
    display(f"  Highest in Year: {max_year} ({profit_by_year[col].max():,.2f})")
    display(f"  Lowest in Year: {min_year} ({profit_by_year[col].min():,.2f})")

# Calculate the average 'tarif_pph_normal' across all years
average_tarif_pph_normal = df_ts['tarif_pph_normal'].mean()
display(f"\nAverage 'tarif_pph_normal' across all years: {average_tarif_pph_normal:.4f}")

# Determine the total 'pph_badan_normal' paid across all years
total_pph_badan_normal = df_ts['pph_badan_normal'].sum()
display(f"Total 'pph_badan_normal' paid across all years: {total_pph_badan_normal:,.2f}")
```

**Reasoning**:
Analyze the relationship between 'pendapatan_operasional' and 'total_hpp' over time by plotting them against the year.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by year and calculate the sum of 'pendapatan_operasional' and 'total_hpp'
revenue_hpp_by_year = df_ts.groupby(df_ts['tahun'].dt.year)[['pendapatan_operasional', 'total_hpp']].sum()

# Plot the trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=revenue_hpp_by_year, markers=True)
plt.title('Trend of Pendapatan Operasional and Total HPP Over Time')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.grid(True)
plt.xticks(revenue_hpp_by_year.index)
plt.tight_layout()
plt.show()
```
